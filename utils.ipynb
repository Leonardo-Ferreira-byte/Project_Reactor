```python
import numpy as np
```

```python
def get_VABP(x):

    return np.mean(x)

def get_slope(x):
    
    return (x[-2] - x[1])/80

def get_MeABP(VABP, slope):

    return VABP - np.exp(-0.94402 - 0.00865 * (VABP - 32)**0.6667 + 2.9979 * slope**0.333)
```

```python
def celcius_to_farenheit(T):

    return T*1.8 + 32
```

```python
temperatures_celcius = np.array([245, 353, 465, 524, 551, 583, 617])

temperatures_farenheit = celcius_to_farenheit(temperatures_celcius)

VABP = get_VABP(temperatures_farenheit)

slope = get_slope(temperatures_farenheit)

get_MeABP(VABP, slope)/1.8 - 32
```
